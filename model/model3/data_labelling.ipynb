```python
import os
import cv2 
from skimage.feature import hog 
from skimage import exposure 
import matplotlib.pyplot as plt

def extract_hog_features(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    features, hog_image = hog(gray, orientations = 9, pixels_per_cell = (8,8), 
                              cells_per_block = (2, 2), visualize = True, block_norm = "L2-Hys")
    
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range = (0, 10))
    # # Visualize HOG image
    # plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
    # plt.axis('off')
    # plt.show()
   

    return features


# image_path = "C:\\Users\\Gen3r\\Documents\\capstone\\ml_model\\data\\dataset1_extracted\\swipe_left\\gesture_swipe_left_0_1714505490.avi_frame_0.jpg"
# image = cv2.imread(image_path)

# # Check if the image is loaded properly
# if image is not None:
#     hog_features = extract_hog_features(image)
#     print(f"HOG Features: {hog_features}")
# else:
#     print("Failed to load the image. Please verify the path.")


# Function to label data
def label_data(dataset_path):
    labeled_data = []

    if not os.path.exists(dataset_path):
        print("path doesn't exist")
        return -1
    
    # Iterate through each gesture label folder in the dataset
    for gesture_label in os.listdir(dataset_path):
        gesture_folder_path = os.path.join(dataset_path, gesture_label)
        
        print(gesture_label)

        # Iterate through each frame in the gesture label folder
        for frame_file in os.listdir(gesture_folder_path):
            frame_path = os.path.join(gesture_folder_path, frame_file)
            # Load the frame
            image = cv2.imread(frame_path)
            
            # Check if the image is loaded properly
            if image is not None:
                # Extract HOG features
                hog_features = extract_hog_features(image)
                
                # Assign the corresponding gesture label
                label = gesture_label
                
                # Store the features and label
                labeled_data.append((hog_features, label))
            else:
                print(f"Failed to load the image: {frame_path}")
    
    return labeled_data

# Example usage
dataset_path = "C:\\Users\\Gen3r\\Documents\\capstone\\ml_model\\data\\dataset1_extracted"
labeled_data = label_data(dataset_path)

print(labeled_data)
```

```python
import pickle 
# Save labeled data to a file
def save_labeled_data(labeled_data, file_path):
    with open(file_path, 'wb') as file:
        pickle.dump(labeled_data, file)

# Load labeled data from a file
def load_labeled_data(file_path):
    with open(file_path, 'rb') as file:
        labeled_data = pickle.load(file)
    return labeled_data

# Example usage: Save labeled data to a file
file_path = "C:\\Users\\Gen3r\\Documents\\capstone\\ml_model\\model\\model3\\labeled_data.pkl"
save_labeled_data(labeled_data, file_path)

# # Example usage: Load labeled data from a file
# loaded_labeled_data = load_labeled_data(file_path)
# print(loaded_labeled_data)
```
